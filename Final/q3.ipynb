```python
import numpy as np
import matplotlib.pyplot as plt
```

Define constants and rate of change function

```python
kp = 1
km = 1
```

```python
def xStarRate(xt,xStar):
    return kp*(xt-xStar)*(xStar)-km*(xStar)
```

```python
def phos(xt,xStar):
    return kp*(xt-xStar)*(xStar)
```

```python
def dephos(xStar):
    return km*(xStar)
```

```python
xtList = [.2,.6,1,1.4,1.8,2]
```

a. Plot forward and reverse rates for each xt value

```python
for i in xtList:
    input = np.linspace(0,i)
    plt.figure()
    plt.plot(input,phos(i,input))
    plt.plot(input,dephos(input))
    plt.legend(['Phosphorylation','Dephosphorylation'])
    plt.xlabel('xStar')
    plt.ylabel('Rate Magnitude (fwd or rev)')
    plt.title('xt = '+str(i))
```

b. Solve for the steady state for arbitrary xtotal 

```python
import sympy
from sympy.solvers import solve
from sympy import Symbol
```

```python
xt,xStar = sympy.symbols('xt,xStar')
sympy.solve(xStarRate(xt,xStar),xStar)
```

in terms of all general parameters

```python
del kp
del km
```

```python
xt,xStar,kp,km = sympy.symbols('xt,xStar,kp,km')
sympy.solve(xStarRate(xt,xStar),xStar)
```

For xt less than or equal to km/kp, there is only one steady state solution, which is at xStar = 0. The forward and reverse rates are zero in this case, and the state is stable because increasing xStar from zero results in a higher dephosphrylation rate than phoshorylation rate.

For xt > km/kp, there are two steady state values of xStar: xStar = 0, and xStar = xt -km/kp. The former is unstable while the latter is stable.

c. Plot stable steady states vs xt

```python
def xStarSteady(kp,km,xt):
    return(-km/kp + xt)
```

```python
xtVals = np.linspace(1,2)
plt.hlines(y = 0,xmin=0,xmax=1)
plt.plot(xtVals,xStarSteady(1,1,xtVals))
plt.xlabel('xt')
plt.ylabel('xStar steady state')
plt.title('Steady states vs xt')
```

Ultrasenstivity refers to the large change in output in response to a small change in input. This occurs for the input values around xt = km/kp, where bifurcation occurs. Steady-state Xstar = 0 for all xt < km/kp, but a small increase in xt just below km/kp to xt > km/kp will create a nonzero output that increases linearly with xt.
